```python
from generations import *
from plot import *
from em import *
```

```python
covariances = create_long_covariances(5, 0.1, 4)
```

```python
N = 10
data, centers, _ = generate_groups(covariances, groups=N, l=10, r=20)
plot_data(data, centers)
```

Запустим алгоритм при условии, что нам известно сколько тут групп

```python
gm = multiEM(n_components=N,  max_iter=20, reg_covar=1e-6, eps=1e-270)
gm.fit(data, covariances=covariances, verbose=2, n_init=30)
```

```python
plot_result(data, gm)
```

Теперь попробуем оценить число групп, искать будем начиная с 6:

```python
gm = iterativeEM(max_iter=100, eps=1e-270, min_center_distance=5)
gm.n_components = 6
gm.fit(data, covariances=covariances, verbose=2, n_init=100)
```

```python
plot_result(data, gm)
```
